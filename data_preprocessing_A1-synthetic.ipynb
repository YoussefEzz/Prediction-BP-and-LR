```python
#read the information of A1-synthetic.txt and load it into a dataframe to preprocess it
import pandas as pd
import numpy as np


#read the .txt file 
df = pd.read_table('Data/A1-synthetic.txt', delimiter = '\t')
df.head()
```

```python
#print the number of rows and columns
print(df.shape)
print(df.columns)
print(df.columns[0])
print(df[df.columns[0]])
print(min(df[df.columns[0]]))
```

```python
#preprocess input 9 columns to scale it's values from 0 to 1
columns = df.shape[1]
inputcolumns = df.columns[0 : 9]
smin = 0
smax = 1
df_normalized = df
for inp_col in inputcolumns:
    column_values = df[inp_col]
    #print(column_values)
    xmin = min(column_values)
    xmax = max(column_values)
    #print( smin + ((smax - smin) / (xmax - xmin) ) * (df[inp_col] - xmin) )
    df_normalized[inp_col] = np.round(smin + ((smax - smin) / (xmax - xmin) ) * (df[inp_col] - xmin), 4)
print(df_normalized)  
```

```python
#preprocess output 10th column to scale it's values from 0.1 to 0.9
columns = df.shape[1]
outputcolumn = df.columns[9]
smin = 0.1
smax = 0.9

column_values = df[outputcolumn]
xmin = min(column_values)
xmax = max(column_values)
df_normalized[outputcolumn] = np.round(smin + ((smax - smin) / (xmax - xmin) ) * (df[outputcolumn] - xmin), 4)
print(df_normalized) 
```

```python
# Write normalized DataFrame to a table-like format (CSV file)
df_normalized.to_csv('Normalized Data/A1-synthetic_normalized.txt', index=False, sep='\t')
```
